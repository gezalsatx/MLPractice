```python
import pandas as pd
import time
import datetime 
import numpy as np
from bokeh.plotting import figure, show
from bokeh.layouts import column
from bokeh.models import (Circle, ColumnDataSource,
                          Line, Plot, HoverTool)
from numpy import linspace
from datetime import date
from everclear.utils.bokeh_utils import BokehUtils


raw_columns=[
[100,   90,    datetime.date(2010, 1, 15)],
[200,   200,   datetime.date(2010, 1, 15)],
[300,   200,   datetime.date(2010, 1, 15)],
[150,   100,   datetime.date(2010, 1, 15)],
[200,   150,   datetime.date(2010, 1, 15)],
[300,   100,   datetime.date(2010, 1, 15)],
[60,    60,    datetime.date(2010, 1, 15)],
[325,   300,   datetime.date(2010, 1, 15)],
[200,   90,    datetime.date(2010, 1, 15)],
[70,    60,    datetime.date(2010, 1, 15)],
[80,    80,    datetime.date(2010, 1, 15)],
]

df = pd.DataFrame(data=raw_columns, columns=
                       ['AUC','AUR','SD'])


circles_source = ColumnDataSource(
    data = dict(
        AUC   = df['AUC'],
        AUR   = df['AUR'],
        SD  = df['SD']
    )
   )


def make_plot(xname, yname):
    
    bokeh_utils = BokehUtils()
    
    hover_tool = HoverTool(
            tooltips=[
                ('AUC', '@AUC{$0.00}'),
                ('AUR', '@AUR{$0.00}' ),
                ('Product Description', '@PD{%F}'),
                ('Start Date', '@SD{%F}'),
                ('Store Number', '@SN{%F}'),
                ('Zone Name', '@ZN{%F}'),
                ],
            formatters={'@SD': 'datetime'},
            mode='mouse')
    
    plot = figure(title='Average Price < Average Cost',
                           x_axis_label='AUC',
                           y_axis_label='AUR',
                           plot_width=400,
                           plot_height=400,
                           min_border_bottom=60,
                           tools=['pan,wheel_zoom,box_zoom,reset', hover_tool]
                           )

    plot.toolbar.logo = None
    plot.xaxis.formatter = bokeh_utils.small_currency_formatter
    plot.yaxis.formatter = bokeh_utils.small_currency_formatter
        
        
#     circle = Circle(
#         x=xname, y=yname, size=6,
#         fill_color=None, line_color="#B7410E", fill_alpha=0.3
#     )
  
    plot.circle(x='AUC',
                y='AUR', 
                size=6,
                source = circles_source,
                fill_color=None, 
                line_color="#B7410E",
                fill_alpha=0.3)
    
#     plot.toolbar.logo = None  
    
    
    x = np.linspace(-40, 10, 100)
    y = x
    
    plot.line(x,
              y,
              line_dash='solid',
              line_width=1,
              line_color='green',
              legend_label="Cost = Price")
  
 #     plot.segment(x0=0, y0=0,
#        color="#FB8072", line_width=2)
#     plot.ray(x=0, y=0, length=0, angle=45,
#       angle_units="deg", color="#FB8072", line_width=2)
    
    return plot


show(make_plot('AUC', 'AUR'))
```
